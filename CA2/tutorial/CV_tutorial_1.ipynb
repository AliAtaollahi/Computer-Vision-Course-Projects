# Introduction to OpenCV

## Reading and Displaying Images

```python
import cv2
from matplotlib import pyplot as plt

# Read an image from file   
image = cv2.imread('example1.jpg')

# Display the image using OpenCV
cv2.imshow('Displayed Image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Display the image using matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title('Displayed Image using Matplotlib')
plt.show()
```

## Basic Operations on Images (Resize, Crop, Rotate)

```python
# Resize the image
resized_image = cv2.resize(image, (300, 300))

# Crop the image
cropped_image = image[50:200, 50:200]

# Rotate the image
(h, w) = image.shape[:2]
center = (w / 2, h / 2)
matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated_image = cv2.warpAffine(image, matrix, (w, h))

# Display the results
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title('Resized Image')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
plt.title('Cropped Image')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image')

plt.show()
```

# Image Basics

## Understanding Pixels and Image Data Representation

```python
# Get image dimensions
(height, width, channels) = image.shape
print(f"Image Dimensions: {height}x{width} pixels, Channels: {channels}")

# Access pixel values
pixel_value = image[100,100]
print(f"Pixel Value at (100, 100): {pixel_value}")
```


## Color Spaces and Conversions (RGB, HSV, Grayscale)

```python
# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Convert to HSV
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Display the results
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')

plt.subplot(1, 3, 3)
plt.imshow(hsv_image)
plt.title('HSV Image')

plt.show()
```

# Basic Image Operations

## Image Arithmetic (Addition, Subtraction)

```python
# Create two images with the same dimensions
image1 = cv2.imread('example1.jpg')
image2 = cv2.imread('example2.jpg')
image1 = cv2.resize(image1, (500, 500))
image2 = cv2.resize(image2, (500, 500))

# Add the images
added_image = cv2.add(image1, image2)

# Subtract the images
subtracted_image = cv2.subtract(image1, image2)

# Display the results
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(added_image, cv2.COLOR_BGR2RGB))
plt.title('Added Image')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(subtracted_image, cv2.COLOR_BGR2RGB))
plt.title('Subtracted Image')

plt.show()
```

## Bitwise Operations (AND, OR, NOT, XOR)

```python
# Create a blank image and draw shapes
blank_image1 = cv2.rectangle(image.copy(), (150, 150), (350, 350), (255, 255, 255), -1)
blank_image2 = cv2.circle(image.copy(), (250, 250), 100, (255, 255, 255), -1)
# Bitwise AND
bitwise_and = cv2.bitwise_and(blank_image1, blank_image2)

# Bitwise OR
bitwise_or = cv2.bitwise_or(blank_image1, blank_image2)

# Bitwise NOT
bitwise_not = cv2.bitwise_not(blank_image1)

# Bitwise XOR
bitwise_xor = cv2.bitwise_xor(blank_image1, blank_image2)

# Display the results
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(bitwise_and, cv2.COLOR_BGR2RGB))
plt.title('Bitwise AND')

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(bitwise_or, cv2.COLOR_BGR2RGB))
plt.title('Bitwise OR')

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(bitwise_not, cv2.COLOR_BGR2RGB))
plt.title('Bitwise NOT')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(bitwise_xor, cv2.COLOR_BGR2RGB))
plt.title('Bitwise XOR')

plt.show()
```

# Image Enhancement Techniques

## Histogram Equalization

```python
# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(gray_image)

# Display the results
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Grayscale Image')

plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Histogram Equalized Image')

plt.show()
```

## Image Filtering (Smoothing, Sharpening, Blurring)

```python
import numpy as np
image = cv2.resize(image , (667,445))
# Apply Gaussian blur
blurred_image = cv2.GaussianBlur(image, (15, 15), 0)

# Apply sharpening kernel
sharpening_kernel = np.array([[-1, -1, -1],
                              [-1, 9, -1],
                              [-1, -1, -1]])
sharpened_image = cv2.filter2D(image, -1, sharpening_kernel)

# Display the results
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.title('Blurred Image')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB))
plt.title('Sharpened Image')

plt.show()
```

## Edge Detection (Sobel, Canny)

```python
# Apply Sobel edge detection
sobelx = cv2.Sobel(image, 2, 1, 0, ksize=5)
sobely = cv2.Sobel(image, 2, 0, 1, ksize=5)

# Apply Canny edge detection
canny_edges = cv2.Canny(image, 100, 200)

# Display the results
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(sobelx, cv2.COLOR_BGR2RGB))
plt.title('Sobel X')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(sobely, cv2.COLOR_BGR2RGB))
plt.title('Sobel Y')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(canny_edges, cv2.COLOR_BGR2RGB))
plt.title('Canny Edges')

plt.show()
```
